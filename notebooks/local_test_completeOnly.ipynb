# BPI 2012 SASRec preprocessing - 500 cases

Builds a local 500-case sample for VSCode development and saves SASRec plus time-aware preprocessing outputs.

```python
from pathlib import Path
import sys

PROJECT_ROOT = Path.cwd().resolve().parent if Path.cwd().name == 'notebooks' else Path.cwd().resolve()
if str(PROJECT_ROOT) not in sys.path:
    sys.path.append(str(PROJECT_ROOT))

print(PROJECT_ROOT)
```

```python
from src.preprocess_bpi2012 import (
    case_length_distribution,
    load_bpi2012_events,
    prepare_sasrec_dataset,
    sample_cases,
    save_preprocess_result,
    summarize_event_table,
)
```

```python
XES_PATH = PROJECT_ROOT / 'data' / 'BPIC2012' / 'BPI_Challenge_2012.xes'
OUT_DIR = PROJECT_ROOT / 'sample_data' / 'bpi2012_500_cases'
MIN_CASE_LENGTH = 3
N_LOCAL_CASES = 500
```

```python
log_df = load_bpi2012_events(
    xes_path=XES_PATH,
    include_lifecycle=False,
    lifecycle_filter='COMPLETE',
)

summarize_event_table(log_df)
```

```python
log_df.head(5)
```

```python
sample_df = sample_cases(log_df, n_cases=N_LOCAL_CASES, strategy='earliest')
result = prepare_sasrec_dataset(sample_df, min_case_length=MIN_CASE_LENGTH)

summarize_event_table(result.events), result.encoding_summary
```
