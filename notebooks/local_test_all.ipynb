# BPI 2012 SASRec preprocessing - full data

This notebook processes the full BPI 2012 COMPLETE-event log.
It keeps the required XES loading condition and produces SASRec baseline files plus time-aware features.

```python
from pathlib import Path
import sys

PROJECT_ROOT = Path.cwd().resolve().parent if Path.cwd().name == 'notebooks' else Path.cwd().resolve()
if str(PROJECT_ROOT) not in sys.path:
    sys.path.append(str(PROJECT_ROOT))

print(PROJECT_ROOT)
```

```python
from src.preprocess_bpi2012 import (
    add_time_features,
    build_sasrec_interactions,
    case_length_distribution,
    encode_ids,
    filter_short_cases,
    load_bpi2012_events,
    save_event_table,
    save_mapping,
    save_sasrec_interactions,
    save_sasrec_interactions_csv,
    summarize_event_table,
)
```

```python
XES_PATH = PROJECT_ROOT / 'data' / 'BPIC2012' / 'BPI_Challenge_2012.xes'
OUT_DIR = PROJECT_ROOT / 'data' / 'processed' / 'bpi2012_complete_only'
MIN_CASE_LENGTH = 3
```

```python
log_df = load_bpi2012_events(
    xes_path=XES_PATH,
    include_lifecycle=False,
    lifecycle_filter='COMPLETE',
)

summarize_event_table(log_df)
```

```python
filtered_df = filter_short_cases(log_df, min_case_length=MIN_CASE_LENGTH)
time_df = add_time_features(filtered_df)
encoded_df, user_map, item_map, encoding_summary = encode_ids(time_df)
sasrec_interactions = build_sasrec_interactions(encoded_df)

summarize_event_table(filtered_df), encoding_summary
```

```python
paths = {
    'events': save_event_table(filtered_df, OUT_DIR / 'events_complete_only_minlen3.csv'),
    'time_features': save_event_table(encoded_df, OUT_DIR / 'events_encoded_time_features.csv'),
    'user_map': save_mapping(user_map, OUT_DIR / 'user_map.csv'),
    'item_map': save_mapping(item_map, OUT_DIR / 'item_map.csv'),
    'sasrec_txt': save_sasrec_interactions(sasrec_interactions, OUT_DIR / 'sasrec_interactions.txt'),
    'sasrec_csv': save_sasrec_interactions_csv(sasrec_interactions, OUT_DIR / 'sasrec_interactions.csv'),
}
paths
```
